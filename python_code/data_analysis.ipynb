```python
path = '/Users/stevenfelix/Dropbox/DataScience/Projects/PsychologyToday/data/'
data = pd.read_csv(path+'psychologytoday2017-02-17.txt', sep = '\t')
```

```python
data.reset_index
del data['Unnamed: 0']
del data['treatmentapproach'] # this doesn't contain data
del data['treatmentapproachnum'] # this is always 0
del data['sexualitynum'] # few therapists include this
del data['specialtiesnum'] # PT limits to 3 specialities, so most everyone lists 3
del data['file']
data.head()
```

```python
data = data[data.title != 'treatment facility']
data.reset_index(inplace = True)
del data['index']
data.head()
```

```python
exclusions = """Group, Associates, Professionals, Boston, Center, New England, Recovery, 
                Massachusetts, Treatment, Network, Newton, Brighton, Arlington, Watertown, 
                Coaching, Consulting, Cambridge, Somerville, Lexington, Waltham""".split(', ')
exclusions = [string.replace('\n', '').replace(' ', '') for string in exclusions]
to_remove = [any(exclusion in title for exclusion in exclusions) for title in data.name]
```

```python
data[to_remove]
```

```python
to_keep = [not b for b in to_remove]
np.sum(to_keep)
```

```python
data = data[to_keep]
data.reset_index(inplace = True)
del data['index']
print data.shape
data.head()
```

```python
list(data.columns)
```

Numerical variables include:
- years: Number of years practicing; For values <10, specific numbers provided, otherwise a lower bound (e.g., "20+ years")
- graduated: Year graduated
- fee: the average of the range listed (e.g., "$100 - 200" --> $150
- issuesnum: number of "issues" that a provider treats
- mentalhealthnum: number of mental health problems a provider treats
- treatmentorientationnum: number of treatment orientation a provider utilizes

```python
summaries = data.describe()
summaries
```

Our "years" variable is probably best considered ordinal, as the data take on larger ranges among more experienced clinicians (e.g., 15-20 years).

```python
yearVals = set(data.years.dropna(inplace=False))
print "Unique values for number of years in practice:"
yearVals
```

Let's make years ordinal:

```python
labels = [ "{0} - {1}".format(i, i + 9) for i in [0,10,20]]
labels.append('30+')
labels
data['years_ord'] = pd.cut(data.years, [0,10,20,30,100], right=False, labels=labels)
counts = data.groupby('years_ord').size()
```

```python
plt.figure(figsize = [5,5])
counts.plot(kind = 'barh')
```

Although providers' profiles must follow a certain structure, providers are free to choose which pieces of information they want to include in their profiles. For instance, some providers may choose not to list the year they graduated, their fee, or which treatment orientations they use.  Consequently, there is a good deal of 'missing' data.

```python
num_columns = list(summaries.columns) # list of numeric columns

fig = plt.figure(figsize = [20,20])

for i,v in enumerate(num_columns):
    i += 1
    newplot = fig.add_subplot(3,2,i)
    newplot.hist(data[v].dropna())
    newplot.set_title(v.replace("_"," "))
```

```python
num_columns.pop(0) # get rid of 'years
```

```python
fig = plt.figure(figsize = [30,30])

for i,v in enumerate(num_columns):
    i += 1
    plotnum = i
    for i2,v2 in enumerate(num_columns[i:]):
        newplot = fig.add_subplot(4,4,plotnum)
        newplot.plot(data[v],data[v2], "o", alpha = .3)
        newplot.set_ylabel(v2)
        newplot.set_xlabel(v)
        plotnum += 4
        #newplot.xlabel(v)
        #       newplot.set_title(v.r)
```

Looking at these scatterplots, there are not any distinct numerical trends, except maybe that providers that treat more issues also subscribe to more types of treatment orientations. Put another way, providers tend to provide similar numbers of each. This could be a genuine, causal connection, or this could be simply result from a desire to look like a more flexible, diverse clinician. This is the equivalent of casting a wide net.
